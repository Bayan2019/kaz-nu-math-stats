# Random Variable

A **random variable** is a numerical outcome of an experiment.

## Distribution

### Cumulative Distribution Function (`.cdf`)

The **cumulative distribution function (cdf)** of a **random variable** $X$ is defined as the function
$$F(x) = \mathrm{P}(X\leq x)$$

The **survival function** of a **random variable** $X$ is
$$S(x) =  \mathrm{P}(X > x) = 1-F(x).$$

> Note the inequality in the cdf is less than or equal to, which is relevant in discrete data because $\mathrm{P}(X = x) \geq 0$, but not in continuous variables as $\mathrm{P}(X = x) = 0$.

### Quantiles (`.ppf`)

For a probability value $\alpha \in [0, 1]$ the $\alpha\mathrm{th}$ quantile of the distribution with **cdf** $F(\cdot)$ is the point $x_{\alpha}$ such that
$$F(x_{\alpha}) = \alpha.$$

### Probability Mass Function (`.pmf`)

The **probability mass function (pmf)** of a **random variable** $X$ is the function that assigns a probability that $X$ takes a particular value
$$p(x)=\mathrm{P}(X=x).$$
To be a valid **pmf**, a function $p(\cdot)$ must satisfy
- $p(x)\geq0$ for $\forall x \in \mathcal{X}$
- $\sum_{x\in\mathcal{X}}p(x) = 1$

### Probability Density Function (`.pdf`)

A **probability density function (pdf)** is a function, say $f(\cdot)$, associated with a continuous **random variable**, $X$, such that areas under **pdf**s correspond to probabilities for that **random variable**
$$P(X\in A) = \int_{A}f(x)dx$$
To be a valid **pdf**, the function $f(\cdot)$ must satisfy
- $f(x)\geq 0$ for all $x$ 
- $\int_{-\infty}^{\infty}f(x)dx=1$

### Mean or expected value (`.mean`)

The **expected value** or **mean** of a **random variable** is the center of mass of its distribution.

#### Discrete Variables

For a discrete **random variable** $X$ with **pmf** $p(x)$, the **mean** is defined as follows
$$\mathbb{E}(X) = \sum_{x}x p(x),$$
where the sum is taken over all possible values $x$ of $X$.

#### Continuous Variables

For a continuous random variable, $X$, with **pdf** $f(x)$ the **mean** or **expected value**
is defined as follows
$$\mathbb{E} X = \int_{-\infty}^{\infty}x f(x)dx.$$

### Variances (`.var`) and Standard Deviation (`.std`)

The **variance** of a **random variable** is a measure of spread of the distribution of that **random variable**. 
If $X$ is a **random variable** with **mean** $\mu$, the **variance** of $X$ is defined as
$$\mathrm{Var}(X) = \mathbb{E}(X-\mu)^2,$$
which is the expected squared distance from the **mean**.

The **standard deviation** of a **random variable** $X$ is defined as the square root of the **variance**. 
More precisely,
$$\mathrm{sd}(X) = \sqrt{\mathrm{Var}(X)} = \sqrt{\mathbb{E}(X-\mu)^2}$$


### Chebyshev’s in- equality

**Chebyshev’s inequality** states that for any **random variable** $X$ with **mean** $\mu$ and **variance** $\sigma^2$
$$\mathrm{P}(|X-\mu|\geq k \sigma) \leq \frac{1}{k^2}.$$

While the observation does not tell us exactly what the **true mean** is or what the **empirical mean** will be in a future experiment, it does provide a lot of useful information.

## iid random variables

Consider a set of $n$ **independent** **random variables** $X_1, \dots, X_n$ and let $f_i(\cdot)$ for $i\in {1, \dots, n}$ be the **pdf** (or **pmf**) of the $i\mathrm{th}$ **random variable**. 
We say that these variables are **independent and identically distributed (iid)** if they are **independent** and all $f_i(\cdot)$ are equal. 
More precisely, $f_1(\cdot) = ... = f_n(\cdot) = f(\cdot)$ and the joint distribution of the vector $X = (X_1, \dots, X_n)$ can be written as
$$f_{X}(x_1, \dots, x_n) = \prod_{i=1}^n f(x_i)$$

> The **iid** assumption on **random variables** is the default assumption for **random samples** and many of the important theories of biostatistics are founded on assuming that variables are **iid**.

## Covariance and correlation

The **covariance** between two **random variables** $X$ and $Y$ is defined as
$$\mathrm{Cov}(X, Y) = \mathbb{E}\big((X-\mathbb{E}X)(Y-\mathbb{E}Y)\big)$$

One of the problems with **covariances** is that they are expressed in units that combine the units of both variables and cannot be compared across pairs of variables that have different units and are hard to transport from one study to another. 
To circumvent this problem the **correlation** of two **random variables** is defined as
$$
\mathrm{Cor}(X, Y) = \frac{\mathrm{Cov}(X, Y)}{\sqrt{\mathrm{Var}(X)\mathrm{Var}(Y)}}.
$$
