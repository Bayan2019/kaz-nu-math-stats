# Bayes rule

## Inroduction

For continuous **random variables** the **Bayes rule** is
$$f_{Y|X=x}(y) = \frac{f_{X|Y=y}(x)f_Y(y)}{\int f_{X|Y=t}(x)f_Y(t)dt}.$$

If the variable $Y$ is discrete then the **Bayes formula** can be written as
$$f_{Y|X=x}(y) = \frac{f_{X|Y=y}(x)f_Y(y)}{\sum_t f_{X|Y=t}(x)f_Y(t)}.$$

## Diagnostic tests

For convenience, denote by $T+$ and $T−$ the events that the result of a diagnostic test is positive or negative, respectively. 
Let $D+$ and $D−$ be the event that the subject of the test has or does not have the disease, respectively.

The **sensitivity** of the test is defined as $\mathrm{P}(T+ |D+)$
$$
\hat{\mathrm{P}}(T+| D+) = \frac{\mathrm{TP}}{\mathrm{TP}+\mathrm{FN}},
$$
the probability that the test is positive conditional on the fact that the subject actually has the disease.

The **specificity** of the test is defined as $\mathrm{P}(T− |D−)$
$$
\hat{\mathrm{P}}(T-| D-) = \frac{\mathrm{TN}}{\mathrm{TN}+\mathrm{FP}},
$$
the probability that the test is negative conditional on the fact that the subject does not have the disease.

The **positive predictive value (PPV)** of the test is $\mathrm{P}(D+ |T+)$
$$
\hat{\mathrm{P}}(D+| T+) = \frac{\mathrm{TP}}{\mathrm{TP}+\mathrm{FP}},
$$
the probability that the subject has the disease conditional on testing positive. 

The **negative predictive value (NPV)** of the test is $P(D− |T−)$
$$
\hat{\mathrm{P}}(D-| T-) = \frac{\mathrm{TN}}{\mathrm{TN}+\mathrm{FN}},
$$
the probability that the subject does not have the disease conditional on testing negative.

The **prevalence** of the disease is $\mathrm{P}(D+)$
$$
\hat{\mathrm{P}}(D+) = \frac{TP+FN}{TP+FP+TN+FN},
$$
the marginal probability of the disease.

## Likelihood ratios of diagnostic tests

The **diagnostic likelihood ratio (DLR)** of a positive test is
$$
\mathrm{DLR}_{+} = \frac{\mathrm{P}(T+| D+)}{\mathrm{P}(T+| D-)} = \frac{\mathrm{sensitivity}}{1-\mathrm{specificity}}.
$$

$$\hat{\mathrm{DLR}}_{+} = \frac{\mathrm{TP}(\mathrm{FP}+\mathrm{TN})}{\mathrm{FP}(\mathrm{TP}+\mathrm{FN})}$$

The **diagnostic likelihood ratio (DLR)** of a negative test is
$$
\mathrm{DLR}_{-} = \frac{\mathrm{P}(T-| D+)}{\mathrm{P}(T-| D-)} = \frac{1-\mathrm{sensitivity}}{\mathrm{specificity}}.
$$

$$\hat{\mathrm{DLR}}_{-} = \frac{\mathrm{FN}(\mathrm{FP}+\mathrm{TN})}{\mathrm{TN}(\mathrm{TP}+\mathrm{FN})}$$

For any probability $p$ of an event we define the **odds** of that event to be
$$\mathrm{Odds}(p) = \frac{p}{1-p}.$$

$$\mathrm{post}\ \mathrm{test}\ \mathrm{odds}(D+) = \mathrm{DLR}_{+}\times\mathrm{prior}\ \mathrm{test}\ \mathrm{odds}(D+)$$

$$\mathrm{post}\ \mathrm{test}\ \mathrm{odds}(D-) = \mathrm{DLR}_{-}\times\mathrm{prior}\ \mathrm{test}\ \mathrm{odds}(D-)$$

## Examples

```python
from scipy.stats import binom
```

"When the subjects using the test were women who collected and tested their own samples, the overall **sensitivity** was 75\%. **Specificity** was also low, in the range 52\% to 75\%." 

```python
sensitivity = 0.75
specificityLower = 0.52
specificityUpper = 0.75
```

Suppose a subject has a negative test. 
Assume the lower bound for the **specificity**. 
What number is closest to the multiplier of the pre-test odds of pregnancy to obtain the post-test odds of pregnancy given a negative test result?

```python
DLRplus = sensitivity/(1-specificityLower)
DLRMinus = (1-sensitivity)/specificityLower
print(f"DLR+={DLRplus}, DLR-={DLRMinus}")
```

Assume the lower value for the **specificity**. 
Suppose a subject has a negative test and that 30% of women taking pregnancy tests are actually pregnant. 
What number is closest to the probability of pregnancy given a negative test?

```python
m=DLRMinus*0.3/(1-0.3)
print(100*m/(1+m))
```

Suppose a subject has a positive test. 
Assume the lower bound for the **specificity**. 
What number is closest to the multiplier of the pre-test odds of pregnancy to obtain the post-test odds of pregnancy given a positive test result?

```python
DLRplus = sensitivity/(1-specificityLower)
DLRMinus = (1-sensitivity)/specificityLower
print(f"DLR+={DLRplus}, DLR-={DLRMinus}")
```

Assume the lower value for the **specificity**. 
Suppose a subject has a positive test and that 30% of women taking pregnancy tests are actually pregnant. 
What number is closest to the probability of pregnancy given the positive test?

```python
m=DLRplus*0.3/(1-0.3)
print(100*m/(1+m))
```

You meet a person at the bus stop and strike up a conversation. 
In the conversation, the person gives the strange answer that at least one of his two children is a girl. 
Thinking on this, you decide to do a probability calculation. 
Assuming only that genders are **iid** with 50% probability each, what is the chance of a two child family having two girls given the information that at least one is a girl?

```python
p1girl = binom.pmf(1, n=2, p=0.5)
p2girl = binom.pmf(2, n=2, p=0.5)
p2girl/p1girl
```
