# Conditional distribution

## Introduction

Consider $B$ to be an event of non-zero probability, that is $\mathrm{P}(B) > 0$. 
Then the **conditional probability** of an event $A$ given that $B$ has occurred is
$$\mathrm{P} = \frac{\mathrm{P}(A\cap B)}{\mathrm{P}(B)}.$$

Let $f_{X, Y}(x, y)$ be a bivariate **density** or **mass** function for **random variables** $X$ and $Y$. 
Let $f_X(x)$ and $f_Y(y)$ be the associated **marginal** **mass** functions or **densities** disregarding the other variables
$$f_Y(y)=\int f_{X, Y}(x, y)dx \quad\quad \mathrm{or} \quad\quad f_Y(y) = \sum_x f_{X, Y}(x, y),$$
where the integral is used if $X$ is continuous and the sum is used if $X$ is discrete.

Then the **conditional density** or **mass** function of $X$ given that $Y=y$ is
$$f_{X|Y}(x|y) = \frac{f_{X, Y}(x, y)}{f_Y(y)}.$$

Bayesians, have a particularly simple notation, where $f_{X|Y}(x|y)=[x|y]$, $f_{X, Y}(x, y)=[x, y]$ and $f_Y(y)=[y]$. 
Thus, in this decluttered notation the same expression can be written as
$$[x|y] = \frac{[x, y]}{[y]}.$$

## Gamma conditionals

Consider the bivariate **pdf**
$$f_{X, Y}(x, y)=ye^{-xy-y},$$
where $x\geq0, y\geq0$.

```python
from sympy import symbols, simplify, integrate, exp, oo
from spb import plot3d_parametric_surface, PB, plot
```

```python
x, y = symbols('x y', real=True, positive=True)
pdf = y*exp(-x*y-y)
# pdfPlot = plot3d(pdf, (x, 0, 4), (y, 0, 4), show=False, surface_color='green')
# pdfPlot.show()
```

```python
u, v = symbols('u v', real=True, positive=True)

gamma = (u, v, v*exp(-u*v-v), (u, 0, 4), (v, 0, 4), 'gamma')
plane = (u, 2, v, (u, 0, 4), (v, 0, 0.4), 'y=2')

surfaces = plot3d_parametric_surface(gamma, plane, 
    backend=PB, wireframe=True, wf_n2=20,
    legend=False, show=False
)

surfaces.show()
```

```python
pdfY = simplify(integrate(pdf, (x, 0, oo)))
print(simplify(pdf/pdfY))
```
