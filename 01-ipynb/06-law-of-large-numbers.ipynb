# Law of Large Numbers (LLN)

## Introduction

### Convergence

A sequence  $(T_{n})_{n}$ of **random variables** **converges in probability** towards the **random variable** $T$ if
$$
\forall \varepsilon>0: \quad \lim_{n\rightarrow\infty}\mathrm{P}(|T_n-T| > \varepsilon) = 0.
$$

A sequence  $(T_{n})_{n}$ of **random variables** **converges almost surely** towards the **random variable** $T$ if
$$
 \mathrm{P}(\lim_{n\rightarrow\infty}T_n = \theta) = 1.
$$

Consider a sequence of **rvs** $X_n$, $n\geq 1$. 
We say that $X_n$ **converges in distribution** to $X$ if
$$F_n(x)=\mathrm{P}(X_n\leq x) \xrightarrow[n]{} F(x)=\mathrm{P}(X\leq x)\quad \forall x.$$

### Law of Large Numbers (LLN)

Establishing that a random sequence converges to a limit is hard. 
Fortunately, we have a theorem that does all the work for us, called the **Law of Large Numbers (LLN)**. 
The **LLN** states that if $X_1, \dots, X_n$ are **iid** from a population with **mean** $\mu$ and **variance** $\sigma^2$, then
$$
\mathrm{P}(|\bar{X}_n-\mu|\geq \varepsilon)\rightarrow 0,
$$
where $\bar{X}_n=\frac{1}{n}\sum_{i=1}^nX_i$.
