## Sample mean

We have discussed the theoretical **mean** of a theoretical **distribution** and how to do the necessary calculations. 
However, in practice we do not know either the true **mean** or the true **distribution** from which observations are drawn.

Let $X_i$ for $i = 1, \dots, n$ be a collection of **random variables**, each from a distribution with **mean** $\mu$. 
The **sample mean** is defined as
$$\bar{X} = \frac{1}{n}\sum_{i=1}^kX_i.$$

After running the experiment we can calculate the **sample mean**, which will contain information about the **true mean**. 
However, the **true mean** is not available to us, just the observations from the experiment.

While the observation does not tell us exactly what the **true mean** is or what the **empirical mean** will be in a future experiment, it does provide a lot of useful information. 
More precisely, it allows us to build **confidence intervals** for the **mean** of a **random variable**.

```python
sum=0
n=len(x)
for v in x:
    sum += v 

print(x.mean()-sum/n, x.mean())
print(x.sum()-sum, x.sum())
```

## Sample and variation series

A **sample** of size $n$ from a **distribution** $F$ is a set of $n$ **independent and identically distributed (iid)** **random variables** $X_{1},X_{2},...,X_{n}$, each having the probability distribution $F$.

If $X_i$ $i=1, \dots, n$ are **random variables** with **variance** $\sigma^2$ the **sample variance** is
defined as
\begin{equation*}
\frac{1}{n}\sum_{k=1}^n (X_k - \bar{X})^2
\end{equation*}
or 
\begin{equation*}
s_n^2 = \frac{1}{n-1}\sum_{k=1}^n (X_k - \bar{X})^2
\end{equation*}
using $n-1$ in the denominator to correct for bias

```python
print(x.var(ddof=0) - ((x - x.mean())**2).sum()/n, x.var(ddof=0))
```

```python
print(x.std(ddof=0) - np.sqrt(((x - x.mean())**2).sum()/n), x.std(ddof=0))
```

```python
print(x.var(ddof=1) - ((x - x.mean())**2).sum()/(n-1), x.var() - x.var(ddof=1), x.var()) # исправленнное выборочное отклонение
```

```python
print(x.std(ddof=1) - np.sqrt(((x - x.mean())**2).sum()/(n-1)), x.std() - x.std(ddof=1), x.std()) # исправленнное стандартное выборочное отклонение
```

## Sample Covariance

Let us consider the case when we have $n$ pairs of random vectors $(X_i, Y_i), i=1, \dots, n$, from the joint distribution of $(X, Y)$ and have a closer look at the **sample covariance** (or **estimator** of the true **covariance**):
$$\frac{1}{n}\sum^n_{i=1}(X_i-\bar{X})(Y_i-\bar{Y})$$

An easy way to obtain an **unbiased** **estimator** of the **covariance** is to use
$$\frac{1}{n-1}\sum^n_{i=1}(X_i-\bar{X})(Y_i-\bar{Y})$$

> When $n$ is moderate or large the distinction between dividing by $n$ or $n − 1$ is irrelevant, though sometimes, when $n$ is small, some differences exist.

```python
n=len(x)
print(((x-x.mean())*(y-y.mean())).sum()/(n)-x.cov(y, ddof=0), x.cov(y, ddof=0))
```

```python
print(((x-x.mean())*(y-y.mean())).sum()/(n-1)-x.cov(y), x.cov(y, ddof=1)-x.cov(y), x.cov(y))
```

```python
print(((x-x.mean())*(y-y.mean())).sum()/(n*np.sqrt(x.var(ddof=0)*y.var(ddof=0)))-x.corr(y), x.corr(y))
```

```python
print(((x-x.mean())*(y-y.mean())).sum()/((n-1)*np.sqrt(x.var()*y.var()))-x.corr(y), x.corr(y))
```
