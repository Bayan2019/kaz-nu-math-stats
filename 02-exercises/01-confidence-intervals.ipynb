## 2

How large a sample would you need in order to estimate the mean height of women to within 2 cm with 
99\% confidence? Assume the standard deviation is approximately 9 cm.

```python
## Решение

Используем формулу минимального объема выборки:
$$n = \left( \frac{z \cdot \sigma}{E} \right)^2$$

Где:
*   $z$ (критическое значение) для 99% = 2.576
*   $\sigma$ (стандартное отклонение) = 9
*   $E$ (допустимая погрешность) = 2

### Расчет:
1.  Подставляем значения:
    $$n = \left( \frac{2.576 \cdot 9}{2} \right)^2$$
2.  Вычисляем числитель: $2.576 \cdot 9 = 23.184$
3.  Делим на погрешность: $23.184 / 2 = 11.592$
4.  Возводим в квадрат: $11.592^2 \approx 134.37$

Ответ: Так как число людей должно быть целым, округляем всегда в большую сторону. Требуется выборка из 135 человек.
```

## 3

Suppose you want to estimate the mean number of chocolate chips in a certain brand of chocolate chip cookies. How large a sample would you need in order to estimate the mean within 0.5 chocolate chips with 95\% confidence? A pilot study indicates that the standard deviation for the population is around 2.55.

```python
import math

# Входные данные
sigma = 2.55
E = 0.5
z = 1.96  # Константа для 95% уверенности

# Расчет
n = (z * sigma / E)**2
n_rounded = math.ceil(n)

print(f"Для оценки среднего с точностью {E} и уверенностью 95% ")
print(f"необходим объем выборки: {n_rounded}")
```

## 4

Below is a random sample of times for a swimming event.
$$
\begin{split}
154.61 \quad\quad 158.03 \quad\quad 164.22 \quad\quad 165.19 \quad\quad 165.64 \quad\quad 168.62 \quad\quad 170.08 \quad\quad 173.17 \quad\quad 174.48\\
175.62 \quad\quad 175.82 \quad\quad 176.47 \quad\quad 176.58 \quad\quad 177.68 \quad\quad 180.33 \quad\quad 183.63 \quad\quad 185.71 \quad\quad 186.49
\end{split}
$$

- Find a 95\% confidence interval for the mean time for this event.
- How large a sample is needed to find the mean time for the swimming event within 0.2 seconds with 95\% confidence?

```python
import math
import numpy as np
from scipy import stats

# 1. Данные из таблицы
data = [
    154.61, 158.03, 164.22, 165.19, 165.64, 168.62, 170.08, 173.17, 174.48,
    175.62, 175.82, 176.47, 176.58, 177.68, 180.33, 183.63, 185.71, 186.49
]

# Основные показатели
n = len(data)
mean = np.mean(data)
std_dev = np.std(data, ddof=1) # Используем исправленное отклонение для выборки
confidence = 0.95

# --- ПУНКТ 1: Нахождение 95% доверительного интервала ---
# Так как выборка небольшая (18 значений), используем t-распределение
t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)
margin_of_error = t_critical * (std_dev / math.sqrt(n))

confidence_interval = (mean - margin_of_error, mean + margin_of_error)

print(f"1. Доверительный интервал (95%): {confidence_interval}")
print(f"   Среднее: {mean:.2f}, Ошибка: {margin_of_error:.2f}")

# --- ПУНКТ 2: Расчет необходимого объема выборки ---
# Условие: точность (E) = 0.2 секунды, уверенность 95%
E = 0.2
z_score = 1.96 # Стандарт для 95% при расчете объема

n_required = (z_score * std_dev / E)**2
n_final = math.ceil(n_required)

print(f"\n2. Для точности в {E} сек. необходим объем выборки: {n_final}")
```

## 5

Below is a random sample of times for a swimming event.
$$
\begin{split}
139 \quad\quad 169 \quad\quad 165 \quad\quad 133 \quad\quad 164\\
150 \quad\quad 163 \quad\quad 137 \quad\quad 132 \quad\quad 134\\
105 \quad\quad 114 \quad\quad 103 \quad\quad 123 \quad\quad 111
\end{split}
$$

- Find a 95\% confidence interval for the mean weight of this event.
- How large a sample would be needed to estimate the mean weight of these apples with 99\% confidence and margin of error at most 5 grams?

```python
import math
import numpy as np
from scipy import stats

# 1. Данные из таблицы (время/вес)
data = [
    139, 169, 165, 133, 164,
    150, 163, 137, 132, 134,
    105, 114, 103, 123, 111
]

# Основные показатели
n = len(data)
mean = np.mean(data)
std_dev = np.std(data, ddof=1)

# --- ПУНКТ 1: 95% доверительный интервал ---
conf_95 = 0.95
t_crit = stats.t.ppf((1 + conf_95) / 2, df=n-1)
moe_95 = t_crit * (std_dev / math.sqrt(n))

ci_95 = (mean - moe_95, mean + moe_95)

print(f"1. Доверительный интервал (95%): ({ci_95[0]:.2f}, {ci_95[1]:.2f})")

# --- ПУНКТ 2: Объем выборки для 99% доверия и ошибки 5 грамм ---
conf_99 = 0.99
E = 5
# z-score для 99% составляет примерно 2.576
z_99 = stats.norm.ppf((1 + conf_99) / 2)

n_req = (z_99 * std_dev / E)**2
n_final = math.ceil(n_req)

print(f"2. Для ошибки {E}г и доверия 99% нужен объем выборки: {n_final}")
```

## 6

A school nurse wants to determine the average caloric content of the snacks that students are choosing. Below is a random sample of the number of Calories in 10 students' snacks.

$$110 \quad\quad 125 \quad\quad 155 \quad\quad 185 \quad\quad 190 \quad\quad 195 \quad\quad 210 \quad\quad 235 \quad\quad 245 \quad\quad 350$$

- Find a 90\% confidence interval for the mean number of calories in the snacks students choose.
- If the nurse wanted to get a better estimate, how large a sample would be needed in order to estimate to within 10 Calories the mean number of calories with 95\% confidence?

```python
import math
import numpy as np
from scipy import stats

# 1. Данные по калориям из условия
calories = [110, 125, 155, 185, 190, 195, 210, 235, 245, 350]

# Основные показатели
n = len(calories)
mean = np.mean(calories)
std_dev = np.std(calories, ddof=1)

# --- ПУНКТ 1: 90% доверительный интервал ---
conf_90 = 0.90
t_crit = stats.t.ppf((1 + conf_90) / 2, df=n-1)
moe_90 = t_crit * (std_dev / math.sqrt(n))

ci_90 = (mean - moe_90, mean + moe_90)

print(f"1. Доверительный интервал (90%): ({ci_90[0]:.2f}, {ci_90[1]:.2f})")
print(f"   Среднее калорий: {mean:.2f}")

# --- ПУНКТ 2: Объем выборки для 95% доверия и ошибки 10 калорий ---
conf_95 = 0.95
E = 10
z_95 = stats.norm.ppf((1 + conf_95) / 2)

n_req = (z_95 * std_dev / E)**2
n_final = math.ceil(n_req)

print(f"\n2. Для точности в {E} калорий и доверия 95% ")
print(f"   необходим объем выборки: {n_final}") 
```
