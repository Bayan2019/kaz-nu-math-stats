## Confidence Intervals

## 1

For a 90\% confidence interval with 10 degrees of freedom, find $t_{\alpha/2}$.

```python
from scipy import stats

# Параметры задачи: 90% доверительный интервал и 10 степеней свободы
alpha = 0.1
df = 10

# Находим t-критическое (для двухстороннего интервала берем 1 - alpha/2)
t_alpha_2 = stats.t.ppf(1 - alpha/2, df)

print(f"Критическое значение t_alpha/2 для df={df}: {t_alpha_2:.3f}")
```

## 4

Below is a random sample of times for a swimming event.
$$
\begin{split}
154.61 \quad\quad 158.03 \quad\quad 164.22 \quad\quad 165.19 \quad\quad 165.64 \quad\quad 168.62 \quad\quad 170.08 \quad\quad 173.17 \quad\quad 174.48\\
175.62 \quad\quad 175.82 \quad\quad 176.47 \quad\quad 176.58 \quad\quad 177.68 \quad\quad 180.33 \quad\quad 183.63 \quad\quad 185.71 \quad\quad 186.49
\end{split}
$$

```python
import numpy as np
from scipy import stats
import math

# Данные из условия
data = [
    154.61, 158.03, 164.22, 165.19, 165.64, 168.62, 170.08, 173.17, 174.48,
    175.62, 175.82, 176.47, 176.58, 177.68, 180.33, 183.63, 185.71, 186.49
]

# 1. Поиск 95% доверительного интервала
n = len(data)
mean = np.mean(data)
std_err = stats.sem(data) # Стандартная ошибка среднего
h = std_err * stats.t.ppf((1 + 0.95) / 2, n - 1) # Погрешность

print(f"1. 95% Доверительный интервал: ({mean - h:.2f}, {mean + h:.2f})")

# 2. Поиск размера выборки для точности 0.2 сек
E = 0.2
sigma = np.std(data, ddof=1) # Используем отклонение из текущей выборки
z = stats.norm.ppf(1 - (1 - 0.95) / 2)
n_new = (z * sigma / E)**2

print(f"2. Необходимый размер выборки (E=0.2): {math.ceil(n_new)}")
```

- Find a 95\% confidence interval for the mean time for this event.
- How large a sample is needed to find the mean time for the swimming event within 0.2 seconds with 95\% confidence?

## 5

Below is a random sample of times for a swimming event.
$$
\begin{split}
139 \quad\quad 169 \quad\quad 165 \quad\quad 133 \quad\quad 164\\
150 \quad\quad 163 \quad\quad 137 \quad\quad 132 \quad\quad 134\\
105 \quad\quad 114 \quad\quad 103 \quad\quad 123 \quad\quad 111
\end{split}
$$

- Find a 95\% confidence interval for the mean weight of this event.
- How large a sample would be needed to estimate the mean weight of these apples with 99\% confidence and margin of error at most 5 grams?

```python
import numpy as np
from scipy import stats
import math

# Данные из условия (вес яблок)
apples = [
    139, 169, 165, 133, 164,
    150, 163, 137, 132, 134,
    105, 114, 103, 123, 111
]

# 1. 95% Доверительный интервал
n = len(apples)
mean = np.mean(apples)
std_err = stats.sem(apples)
h = std_err * stats.t.ppf((1 + 0.95) / 2, n - 1)

print(f"1. 95% Доверительный интервал для веса: ({mean - h:.2f}, {mean + h:.2f})")

# 2. Размер выборки для точности 5 грамм и 99% уверенности
E = 5
sigma = np.std(apples, ddof=1)
z = stats.norm.ppf(1 - (1 - 0.99) / 2)
n_new = (z * sigma / E)**2

print(f"2. Необходимый размер выборки (E=5, 99%): {math.ceil(n_new)}")
```

## 6

A school nurse wants to determine the average caloric content of the snacks that students are choosing. Below is a random sample of the number of Calories in 10 students' snacks.

$$110 \quad\quad 125 \quad\quad 155 \quad\quad 185 \quad\quad 190 \quad\quad 195 \quad\quad 210 \quad\quad 235 \quad\quad 245 \quad\quad 350$$

- Find a 90\% confidence interval for the mean number of calories in the snacks students choose.
- If the nurse wanted to get a better estimate, how large a sample would be needed in order to estimate to within 10 Calories the mean number of calories with 95\% confidence?

```python
import numpy as np
from scipy import stats
import math

# Данные из условия (калории)
calories = [110, 125, 155, 185, 190, 195, 210, 235, 245, 350]

# 1. 90% Доверительный интервал (используем t-распределение, так как n=10)
n = len(calories)
mean = np.mean(calories)
std_err = stats.sem(calories)
# Квантиль для 90% интервала
h = std_err * stats.t.ppf((1 + 0.90) / 2, n - 1)

print(f"1. 90% Доверительный интервал для калорий: ({mean - h:.2f}, {mean + h:.2f})")

# 2. Размер выборки для точности 10 калорий и 95% уверенности
E = 10
sigma = np.std(calories, ddof=1)
z = stats.norm.ppf(1 - (1 - 0.95) / 2)
n_new = (z * sigma / E)**2

print(f"2. Необходимый размер выборки (E=10, 95%): {math.ceil(n_new)}")
```
